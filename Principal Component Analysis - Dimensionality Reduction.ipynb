```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("Bank-data.csv")
dataset.head()
```

```python
dataset['Churn'] = dataset['Churn'].astype("category")
dataset['Churn'] = dataset['Churn'].cat.codes
dataset.head()
```

```python
dataset['Churn'].value_counts()
```

```python
# split
x = dataset.iloc[:,0:-1].values
y = dataset.iloc[:,-1].values
```

```python
x
```

```python
# splitting the data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.75, random_state=0)
```

```python
# Building Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
```

```python
y_pred = logmodel.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
print(confusion_matrix(y_test, y_pred))\

print(accuracy_score(y_test, y_pred))
```

```python
# Logistic Regression Accuracy - 90.7%
```

```python
# PCA : n_components = 2
pca = PCA(n_components=2, random_state=0) # None - all componant / variables/values 
# none - by default system will provide the slice count
x_train_2 = pca.fit_transform(x_train)
x_test_2 = pca.fit_transform(x_test)
```

```python
print(x_train_2)
```

```python
# Logistic Regression with PCA method
logmodel_pca = LogisticRegression()
logmodel_pca.fit(x_train_2, y_train)
```

```python
y_pred_pca = logmodel_pca.predict(x_test_2)
```

```python
accuracy_score(y_test, y_pred_pca)
```

```python
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(estimator=logmodel_pca, X = x_train_2, y = y_train, cv=10)
accuracy
```
