```python
def stats(x):
    x = pd.Series(x)
    Min = x.min()
    Max = x.max()
    Mean = x.mean()
    res = [Min,Max,Mean]
    indx = ["Min","Max","Mean"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = new_data.apply(stats)
loc_stats
```

```python
day_stats = new_data.apply(stats,axis=1)
day_stats.head()
```
