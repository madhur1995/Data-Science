```python
import pandas as pd
import seaborn as sns
from numpy import median, mean
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/athlete_events.csv')
```

```python
df
```

```python
g = sns.catplot('Sport', data=df, kind="count")
```

```python
olympics_winners = df.dropna(subset=['Medal'])
olympics_winners.head()
```

```python
olympics_winners.shape
```

```python
olympics_winners_2016 = olympics_winners[(olympics_winners.Year == 2016)]
```

```python
olympics_winners_2016
```

```python
olympics_winners_2016.shape
```

```python
olympics_winners_2016['Sport'].value_counts()
```

```python
sns.countplot(x="Sport", data=olympics_winners_2016)
```

```python
# list the top 5 sports
top_sports = ['Athletics', 'Swimming', 'Rowing', 'Football', 'Hockey']

# subset the dataframe to include data from the top sports
olympics_top_sports_winners_2016 = olympics_winners_2016[(olympics_winners_2016.Sport.isin(top_sports))]
olympics_top_sports_winners_2016.head()
```

```python
olympics_top_sports_winners_2016.shape
```

```python
sns.countplot(x="Sport", data=olympics_top_sports_winners_2016)
```

```python
g = sns.catplot('Sport', data=olympics_top_sports_winners_2016, kind="count")
```

```python
g = sns.catplot('Age', data=olympics_top_sports_winners_2016, kind="box")
```

```python
g = sns.catplot('Height', data=olympics_top_sports_winners_2016, kind="violin")
```
