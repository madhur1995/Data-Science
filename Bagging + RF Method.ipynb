```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
daib = pd.read_csv("M4_Diabetes.csv")
daib.head(20)
```

```python
daib.isnull().sum()
```

```python
daib['Is_Diabetic'] = daib['Is_Diabetic'].astype("category")
daib['Is_Diabetic'] = daib['Is_Diabetic'].cat.codes
daib.head(10)
```

```python
daib.info()
```

```python
daib['Is_Diabetic'].value_counts()
```

```python
x = daib.iloc[:,0:8]
y = daib.iloc[:,8]

print(x)
print(y)
```

```python
# Splitting the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Bagging concept
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier()
bag_model.fit(x_train, y_train)
```

```python
y_pred_bag_train = bag_model.predict(x_train)
y_pred_bag_test = bag_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_train,y_pred_bag_train))
print(accuracy_score(y_test,y_pred_bag_test))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred_rf_train = rf.predict(x_train)
y_pred_rf_test = rf.predict(x_test)
```

```python
print(accuracy_score(y_train,y_pred_rf_train))
print(accuracy_score(y_test,y_pred_rf_test))
```

```python
# Cross Validation Method
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(estimator=bag_model, X=x_train, y=y_train, cv=15)
accuracy
```

```python
# Cross Validation Method
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(estimator=rf, X=x_train, y=y_train, cv=20)
accuracy
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred_lr_train = lr.predict(x_train)
y_pred_lr_test = lr.predict(x_test)
```

```python
print(accuracy_score(y_train,y_pred_lr_train))
print(accuracy_score(y_test,y_pred_lr_test ))
```
